# 한국타이어 POC

## 데이터 소스

### 1. 배송 실적 데이터
- 파일: `1.업체 공유용_배송 실적_2025년 2월 11월_20260317.csv`
- 내용: 2025년 2월~11월 배송 실적 데이터
- 주요 컬럼:
  - Shipment Making Date, Actual G/I Date: 출하/출고일
  - G/I Plant, Actual Plant: 출발지 공장
  - DC: CDC/RDC 구분
  - Shipment Type, Shipment No., Delivery No.: 수송 유형 및 번호
  - Ship-to party, Shipment Route: 배송처 및 경로
  - Region Code, 광역도시, 시군구: 지역 정보
  - Material: 자재 코드
  - PCR/LTR 구분, Group, Size, Inch: 제품 분류 및 규격
  - Q'ty, Measurement: 수량 및 메저
  - Cost of Shipment, Distance: 운송비 및 거리
  - Type of Truck: 차량 톤수
  - Picking/Loading/Truck/Real G/I 시간: 작업 일시

### 2. 수송 실적 데이터
- 파일: `2.업체 공유용_수송 실적_2025년 2월 11월_20260317.csv`
- 내용: 2025년 2월~11월 수송 실적 데이터
- 주요 컬럼:
  - Origin: 출발지
  - Shipment, Delivery: 수송/배송 번호
  - 출발지, 도착지: 물류창고 코드 및 명칭
  - Material: 자재 코드
  - Delivery quantity: 배송 수량
  - 상품구분: PC/LT, TB 등
  - G/R Date: 입고일
  - 기준욱반비: 운송 비용
  - Ton: 차량 톤수

### 3. 배송 단가표 데이터
- 파일: `3.업체 공유용_배송 단가표_2025년 2월 11월__20260317.csv`
- 내용: 2025년 2월, 11월 기준 배송 단가표
- 구조:
  - 거리 구간별 (1~10km, 11~20km, ..., 381~410km) 운송 단가
  - RDC별 단가: 중부RDC, 평택RDC, 제천RDC, 계룡RDC, 칠곡RDC
  - 차량 톤수별: 1T, 3.5T, 5T, 8T, 11T, 24T
- 용도: 거리 및 차량 톤수별 기준 운송비 산정

### 4. 행정동 경계 데이터 (Shapefile)
- 파일: `map/BND_ADM_DONG_PG.shp`
- 내용: 행정동(읍멘동) 단위 경계 데이터
- 구조:
  - BASE_DATE: 기준일자
  - ADM_CD: 행정동 코드 (10자리)
  - ADM_NM: 행정동명
- 시군구 코드 추출: ADM_CD 앞 5자리
- 용도: 배송 지역 시각화 및 위치 기반 분석

### 5. 시군구 경계 데이터 (Shapefile)
- 파일: `map/BND_SIGUNGU_PG.shp`
- 내용: 시군구 단위 경계 데이터
- 구조:
  - BASE_DATE: 기준일자
  - SIGUNGU_CD: 시군구 코드 (5자리)
  - SIGUNGU_NM: 시군구명
- 용도: 시군구 단위 집계 및 시각화
- 배송 실적의 '시군구' 컬럼과 직접 매칭 가능

### 6. 배송지 마스터 데이터
- 파일: `00_draft/배송지_마스터.csv`
- 내용: 배송지(Ship-to party)와 RDC 담당 매핑 정보
- 주요 컬럼:
  - Ship-to party: 배송지 코드
  - Shipment Route: 배송 경로 코드
  - Region Code: 지역 코드
  - 광역도시, 시군구: 지역 정보
  - 담당 RDC 코드, 담당 RDC 명: 담당 물류센터

### 7. 배송지코드 지역 매핑 데이터
- 파일: `00_draft/배송지코드_지역매핑.csv`
- 내용: Region Code와 광역도시/시군구 매핑
- 주요 컬럼:
  - 배송지코드: Region Code
  - 광역도시: 시/도 단위 지역
  - 시군구: 시/군/구 단위 지역

### 8. 물류센터 위치 데이터
- 파일: `00_draft/물류센터_위치.csv`
- 내용: RDC(물류센터)의 위치 정보
- 주요 컬럼:
  - Plant코드: RDC 코드 (11RC, 11R2, 11R9, 11RD, 11RE)
  - 물류센터: RDC 이름
  - 주소: 상세 주소
  - 위도(Y), 경도(X): 좌표 정보

## RDC별 담당 현황 분석

```python
# 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from utils import (
    load_delivery_master, 
    load_rdc_locations, 
    load_sigungu_shapefile
)

# 한글 폰트 설정
plt.rcParams['font.family'] = ['AppleGothic', 'Malgun Gothic', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 로드
delivery_master = load_delivery_master()
rdc_locations = load_rdc_locations()
sigungu_gdf = load_sigungu_shapefile()

print('배송지 마스터 샘플:')
print(delivery_master.head())
print()
print('물류센터 위치:')
print(rdc_locations)
```

```python
# RDC별 통계 계산
# 1. 배송지 수
rdc_stats = delivery_master.groupby(['담당 RDC 코드', '담당 RDC 명']).agg({
    'Ship-to party': 'nunique',
    '시군구': 'nunique',
    '광역도시': 'nunique'
}).reset_index()

rdc_stats.columns = ['RDC 코드', 'RDC 명', '배송지 수', '시군구 수', '광역도시 수']
rdc_stats = rdc_stats.sort_values('배송지 수', ascending=False)

print('RDC별 담당 현황:')
print(rdc_stats)
```

```python
# 2. RDC별 담당 현황 시각화 (그래프)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 배송지 수
axes[0].bar(rdc_stats['RDC 명'], rdc_stats['배송지 수'], color='steelblue')
axes[0].set_title('RDC별 담당 배송지 수', fontsize=14, fontweight='bold')
axes[0].set_ylabel('배송지 수')
axes[0].tick_params(axis='x', rotation=45)

# 시군구 수
axes[1].bar(rdc_stats['RDC 명'], rdc_stats['시군구 수'], color='coral')
axes[1].set_title('RDC별 담당 시군구 수', fontsize=14, fontweight='bold')
axes[1].set_ylabel('시군구 수')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 3. RDC별 담당 시군구 목록 추출
rdc_sigungu = delivery_master.groupby(['담당 RDC 코드', '담당 RDC 명', '광역도시'])['시군구'].unique().reset_index()
rdc_sigungu['시군구 수'] = rdc_sigungu['시군구'].apply(len)

for _, row in rdc_sigungu.iterrows():
    print(f"\n{row['담당 RDC 명']} ({row['광역도시']}): {row['시군구 수']}개 시군구")
    print(f"  {', '.join(row['시군구'][:5])}{'...' if len(row['시군구']) > 5 else ''}")
```

## 지도 시각화: RDC 위치와 담당 시군구

```python
# RDC 위치를 GeoDataFrame으로 변환
from shapely.geometry import Point

rdc_locations['geometry'] = rdc_locations.apply(
    lambda row: Point(row['경도(X)'], row['위도(Y)']), axis=1
)
rdc_gdf = gpd.GeoDataFrame(rdc_locations, geometry='geometry', crs='EPSG:4326')

print('RDC GeoDataFrame:')
print(rdc_gdf[['물류센터', '주소', '위도(Y)', '경도(X)']])
```

```python
# RDC별 담당 시군구 GeoDataFrame 생성
# 배송지 마스터의 시군구와 시군구 shapefile 매칭

# 시군구명 정제 함수
def normalize_sigungu_name(name):
    # '창원시 마산합포구' → '창원시마산합포구' 또는 '창원시'
    name = str(name).strip()
    if '창원시' in name:
        return '창원시'
    if '청주시' in name:
        return '청주시'
    if '전주시' in name:
        return '전주시'
    if '고양시' in name:
        return '고양시'
    if '성남시' in name:
        return '성남시'
    if '수원시' in name:
        return '수원시'
    if '안산시' in name:
        return '안산시'
    if '안양시' in name:
        return '안양시'
    if '용인시' in name:
        return '용인시'
    if '부천시' in name:
        return '부천시'
    if '평택시' in name:
        return '평택시'
    if '제주시' in name:
        return '제주시'
    if '서귀포시' in name:
        return '서귀포시'
    return name

# 시군구 GeoDataFrame에 정제된 이름 추가
sigungu_gdf['SIGUNGU_NM_CLEAN'] = sigungu_gdf['SIGUNGU_NM'].apply(normalize_sigungu_name)

# 배송지 마스터에서 RDC별 시군구 목록 추출
rdc_sigungu_list = delivery_master.groupby('담당 RDC 코드')['시군구'].apply(
    lambda x: list(set([normalize_sigungu_name(s) for s in x]))
).to_dict()

print('RDC별 담당 시군구 (정제된 이름):')
for rdc, sigungu_list in rdc_sigungu_list.items():
    print(f'{rdc}: {len(sigungu_list)}개 시군구')
```

```python
# RDC별 색상 정의
rdc_colors = {
    '11R2': '#FF6B6B',  # 평택RDC - 빨강
    '11RC': '#4ECDC4',  # 계룡RDC - 청록
    '11RD': '#45B7D1',  # 칠곡RDC - 파랑
    '11R9': '#96CEB4',  # 제천RDC - 초록
    '11RE': '#FFEAA7',  # 중부RDC - 노랑
    '11RB': '#DDA0DD'   # 제주RDC - 복숭아
}

rdc_names = {
    '11R2': '평택RDC',
    '11RC': '계룡RDC',
    '11RD': '칠곡RDC',
    '11R9': '제천RDC',
    '11RE': '중부RDC',
    '11RB': '제주RDC'
}

# 전국 지도에 RDC별 담당 시군구 표시
fig, ax = plt.subplots(figsize=(16, 12))

# 기본 시군구 경계 그리기 (회색)
sigungu_gdf.boundary.plot(ax=ax, color='lightgray', linewidth=0.5, alpha=0.5)

# RDC별 담당 시군구 색상으로 표시
for rdc_code, sigungu_list in rdc_sigungu_list.items():
    color = rdc_colors.get(rdc_code, 'gray')
    
    # 해당 RDC의 시군구 필터링
    mask = sigungu_gdf['SIGUNGU_NM_CLEAN'].isin(sigungu_list)
    rdc_sigungu_gdf = sigungu_gdf[mask]
    
    if len(rdc_sigungu_gdf) > 0:
        rdc_sigungu_gdf.plot(ax=ax, color=color, alpha=0.6, label=rdc_names.get(rdc_code, rdc_code))

# RDC 위치 표시 (검은 별표)
rdc_gdf.plot(ax=ax, color='black', marker='*', markersize=300, label='RDC 위치')

# RDC 이름 표시
for idx, row in rdc_gdf.iterrows():
    ax.annotate(
        row['물류센터'],
        xy=(row.geometry.x, row.geometry.y),
        xytext=(5, 5),
        textcoords='offset points',
        fontsize=10,
        fontweight='bold',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8)
    )

ax.set_title('RDC별 담당 시군구 및 RDC 위치', fontsize=16, fontweight='bold', pad=20)
ax.legend(loc='upper right', title='RDC')
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 개별 RDC별 상세 지도 (각 RDC별로 하나씩)
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for idx, (rdc_code, sigungu_list) in enumerate(rdc_sigungu_list.items()):
    ax = axes[idx]
    color = rdc_colors.get(rdc_code, 'gray')
    
    # 해당 RDC의 시군구
    mask = sigungu_gdf['SIGUNGU_NM_CLEAN'].isin(sigungu_list)
    rdc_sigungu_gdf = sigungu_gdf[mask]
    
    # 전국 경계 (연한 회색)
    sigungu_gdf.boundary.plot(ax=ax, color='lightgray', linewidth=0.3, alpha=0.3)
    
    # 담당 시군구 (색상으로 표시)
    if len(rdc_sigungu_gdf) > 0:
        rdc_sigungu_gdf.plot(ax=ax, color=color, alpha=0.7)
        
        # 해당 지역의 경계 박스 계산하여 줌인
        bounds = rdc_sigungu_gdf.total_bounds
        ax.set_xlim(bounds[0] - 0.5, bounds[2] + 0.5)
        ax.set_ylim(bounds[1] - 0.5, bounds[3] + 0.5)
    
    # RDC 위치 표시
    rdc_point = rdc_gdf[rdc_gdf['Plant코드'] == rdc_code]
    if len(rdc_point) > 0:
        rdc_point.plot(ax=ax, color='red', marker='*', markersize=200)
    
    ax.set_title(f"{rdc_names.get(rdc_code, rdc_code)}\n담당 {len(sigungu_list)}개 시군구", fontweight='bold')
    ax.axis('off')

# 빈 subplot 숨기기
if len(rdc_sigungu_list) < 6:
    for idx in range(len(rdc_sigungu_list), 6):
        axes[idx].axis('off')

plt.suptitle('RDC별 담당 시군구 상세 지도', fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```
